# 二：基于Attention的seq2seq

##### 1、Encoder的每一步的输出是一个encoder_outputs, 用encoder_outputs 和 decoder 前一步传来的 hidden_state做比较 计算权值. 
##### 2、用注意力权重对每一步的encoder_outputs权值相加, 得到 context_vector 
##### 3、拿计算得到的context_vector和此时decoder的输入, 作为decoder的输入. 

##### Bahdanau注意力计算： 将encoder所有步的输出EO经过一个全连接层的结果 和 H经过一个全连接层的结果 相加，送入tanh激活函数，再经过一个全连接网络
